```python
import research as r
import scipy
import pandas

varpath = "/Users/dk/research/src/himalaya/var/"
limit = 1000000
B = [("f101","min_fitness"),
    ("griewangk", "min_fitness"),
    ("n32k8", "max_fitness"),
    ("rana", "min_fitness"),
    ("rosen", "min_fitness")]
    
def load(name, bench, files):
    D = r.load_files(r.find_files(varpath + "/" + name, bench[0]+files))
    D["treatment"] = name
    D["benchmark"] = bench[0]
    D["fitness"] = D[bench[1]]
    return D[D['evaluation']<=limit]

def load_final(name, bench, files):
    D = load(name, bench, files)
    return D[D['evaluation']==limit]

def ciplot(name, bench, files):
    W = load(name, bench, files)
    W = W[W['evaluation']<=1000000]
        
    figure()
    r.quick_ciplot('evaluation', 'fitness', W)
    title(name + ": " + bench[0])
    xlabel("evaluation")
    ylabel("fitness")
    return W
    
def for_benchmarks(benchmarks, treatments, summary):
    R = pandas.DataFrame()
    for b in benchmarks:
        for t in treatments:
            R = R.append(summary(t,b,".*/fitness_evaluations.dat"))
    return R
```

```python
D_all = for_benchmarks(B, ["001-alps", "002-qhfc", "005-qhfc", "003-himalaya", "004-himalaya"], load)
D_final = D_all[D_all['evaluation']==limit]
```

```python
for i,g in D_all.groupby("benchmark"):
    figure()
    lines = []
    labels = []
    for j,h in g.groupby("treatment"):
        lines.append(r.quick_ciplot('evaluation', 'fitness', h))
        labels.append(j)
    title(i)
    legend(lines, labels, 'upper left')
    xlim(0,limit)
    xlabel("evaluation")
    ylabel("fitness")
```
